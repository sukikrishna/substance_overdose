```python
import os
folder_path = "seed42"
```

```python
path = os.path.join(folder_path, 'hyper_csv_' + folder_path + '.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv(path)

# Clean column names by removing any extra spaces
df.columns = df.columns.str.strip()

# Convert columns to appropriate types if necessary (e.g., Lookback Period to numeric)
df['Lookback Period'] = pd.to_numeric(df['Lookback Period'], errors='coerce')

# Set up the plot style
sns.set(style="whitegrid")

# Plot RMSE vs Lookback Period
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Lookback Period', y='LSTM RMSE', label='LSTM RMSE', marker='o', color='b')
sns.lineplot(data=df, x='Lookback Period', y='SARIMA RMSE', label='SARIMA RMSE', marker='o', color='r')
plt.title('RMSE vs Lookback Period')
plt.xlabel('Lookback Period')
plt.ylabel('RMSE')
plt.legend()
plt.show()

# Plot RMSE Standard Deviation vs Lookback Period
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Lookback Period', y='LSTM RMSE Std', label='LSTM RMSE Std', marker='o', color='b')
sns.lineplot(data=df, x='Lookback Period', y='SARIMA RMSE Std', label='SARIMA RMSE Std', marker='o', color='r')
plt.title('RMSE Std vs Lookback Period')
plt.xlabel('Lookback Period')
plt.ylabel('RMSE Std')
plt.legend()
plt.show()

# Plot MAPE vs Lookback Period
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Lookback Period', y='LSTM MAPE', label='LSTM MAPE', marker='o', color='b')
sns.lineplot(data=df, x='Lookback Period', y='SARIMA MAPE', label='SARIMA MAPE', marker='o', color='r')
plt.title('MAPE vs Lookback Period')
plt.xlabel('Lookback Period')
plt.ylabel('MAPE')
plt.legend()
plt.show()

# Plot MAPE Standard Deviation vs Lookback Period
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Lookback Period', y='LSTM MAPE Std', label='LSTM MAPE Std', marker='o', color='b')
sns.lineplot(data=df, x='Lookback Period', y='SARIMA MAPE Std', label='SARIMA MAPE Std', marker='o', color='r')
plt.title('MAPE Std vs Lookback Period')
plt.xlabel('Lookback Period')
plt.ylabel('MAPE Std')
plt.legend()
plt.show()

# Scatter plot of RMSE vs RMSE Std
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='LSTM RMSE', y='LSTM RMSE Std', label='LSTM RMSE Std', color='b')
sns.scatterplot(data=df, x='SARIMA RMSE', y='SARIMA RMSE Std', label='SARIMA RMSE Std', color='r')
plt.title('RMSE vs RMSE Std')
plt.xlabel('RMSE')
plt.ylabel('RMSE Std')
plt.legend()
plt.show()

# Scatter plot of MAPE vs MAPE Std
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='LSTM MAPE', y='LSTM MAPE Std', label='LSTM MAPE Std', color='b')
sns.scatterplot(data=df, x='SARIMA MAPE', y='SARIMA MAPE Std', label='SARIMA MAPE Std', color='r')
plt.title('MAPE vs MAPE Std')
plt.xlabel('MAPE')
plt.ylabel('MAPE Std')
plt.legend()
plt.show()
```
